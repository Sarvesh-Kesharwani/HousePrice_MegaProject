#### converting all 3 area_cols from sqm to sqft

```python
pd.set_option('display.max_colwidth', 1000)
data[['areaWithType', 'SuperBuiltupArea_sqft',	'BuiltupArea_sqft',	'CarpetArea_sqft']].sample(5)
```

```python
import re
def convert_to_sqft(area_text, area_value):
    # basically we are looking builtup/superbuiltup/carpet area in the areaWithType col and if the extracted area is in sq.m. format
    # we are converting it to sqft unit. otherwise if there is no match, it is already in sqft unit. 
    if not area_value:
        return None
    match = re.search(r"{} \((\d+\.?\d*) sq.m.\)*".format(area_value), str(area_text))
    if match:
        sqm_value = float(match.group(1))
        sqft_value = sqm_value*10.76391
        return sqft_value
    return area_value
```

```python
data['CarpetArea_sqft'] = data.apply(lambda x: convert_to_sqft(x['areaWithType'], x['CarpetArea_sqft']), axis=1)
data['BuiltupArea_sqft'] = data.apply(lambda x: convert_to_sqft(x['areaWithType'], x['BuiltupArea_sqft']), axis=1)
data['SuperBuiltupArea_sqft'] = data.apply(lambda x: convert_to_sqft(x['areaWithType'], x['SuperBuiltupArea_sqft']), axis=1)
```
