### loading data

```python
import pandas as pd
data = pd.read_csv(r'..\3. eda_n_feature engineering\2.1 area_feature FEed.csv')
data.head()
```

### checking kind of values present in this col

```python
pd.set_option('display.max_colwidth', 1000)
data['additionalRoom']
```

```python
data['additionalRoom'] = data['additionalRoom'].str.strip().str.lower()
data['additionalRoom']
```

```python
x = data['additionalRoom'].value_counts(dropna=False)
x[x>15]
```

```python
# we can see they are not too many categories, we can even list them here:
    # not available
    # servant room
    # study room
    # pooja room
    # store room
    # study room
    # others
```

### 

```python
data['servantRoom'] = data['additionalRoom'].fillna(' ').str.contains('servant')
data['servantRoom']

data['studyRoom'] = data['additionalRoom'].fillna(' ').str.contains('study')
data['studyRoom']

data['poojaRoom'] = data['additionalRoom'].fillna(' ').str.contains('pooja')
data['poojaRoom']

data['storeRoom'] = data['additionalRoom'].fillna(' ').str.contains('store')
data['storeRoom']

data['studyRoom'] = data['additionalRoom'].fillna(' ').str.contains('study')
data['studyRoom']

data['otherAdditionalRoom'] = data['additionalRoom'].fillna(' ').str.contains('other')
data['otherAdditionalRoom']
```

```python
# not available - none of the below 6 col 1
# servant room - servantRoom_col is 1
# study room - studyRoom_col is 1
# pooja room - poojaRoom_col is 1
# store room - storeRoom_col is 1
# study room - studyRoom_col is 1
# others - otherAdditionalRoom_col is 1
```

```python
pd.set_option('display.max_colwidth', 1000)
data.sample(3)
```
