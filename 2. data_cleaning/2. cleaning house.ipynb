```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

data = pd.read_csv(r'F:\Github_dsc\HousePrice_MegaProject\1. data_gathering\3. merging data\cx_data\1. manually cleaned data\independent-house - independent-house.csv')
data.sample(3)
```

```python
data.drop(columns=['link', 'property_id'], inplace=True)
```

```python
data.sample(3)
```

# society col

```python
data['society'].value_counts(dropna=False)
```

```python
import regex as re
data['society'].str.lower().str.strip().astype(str).apply(lambda row: re.sub(r'\d+(\.\d+)?\s?★', '', row)).value_counts(dropna=False)
```

```python
data['society'] = data['society'].str.lower().str.strip().astype(str).apply(lambda row: re.sub(r'\d+(\.\d+)?\s?★', '', row))
```

```python
data.sample(3)
```

```python
data['price'].value_counts()
```

```python
def clean_price_col(row):
    if 'lac' in row:
        return round(float(row.split(' ')[0])/100, 2)
    elif 'crore' in row:
        return round(float(row.split(' ')[0]), 2)
    else:
        return row

data['price'] = data['price'].str.strip().str.lower().astype(str).apply(clean_price_col)
```

```python
data['price'].value_counts(dropna=False)
```

```python
data['price'].value_counts()
```

```python
data = data[data['price']!='price on request']
```

```python
data.rename(columns={'price':'price_in_crores'}, inplace=True)
```

```python
data.sample(3)
```
