## loading data

```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

data = pd.read_csv(r'..\2. data_cleaning\0. manually(WPS) cleaned datasets\flats - flats.csv')
data.sample(3)
```

## dropping unuseful cols

```python
data.drop(columns=['link', 'property_id'], axis=1, inplace=True)
data.sample(3)
```

## cleaning society col

```python
data['society']
```

```python
data['society'].value_counts().shape
```

```python
import re
data['society'] = data['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name).strip().lower()))
```

```python
data['society'].value_counts().shape
```

## cleaning price col

```python
data['price'].value_counts()
```

```python
data[data['price']=='Price on Request'].head(3)
```

```python
data = data[data['price']!='Price on Request']
```

```python
data.sample(3)
```

```python
data['price'].value_counts()
```

```python
def clean_price_col(row):
    if 'crore' in row:
        return round(float(row.split(' ')[0]), 2)
    elif 'lac' in row:
        return round(float(row.split(' ')[0])/100, 2)
    else:
        return row
```

```python
data['price'] = data['price'].str.strip().str.lower().astype('str').apply(clean_price_col)
data['price'].value_counts()
```

```python
data.rename(columns={'price':'price_in_crores'}, inplace=True)
data.sample(3)
```

## cleaning `area` col

```python
# it is actually the rate in per sqft and not the area, so let's fix it.
```

```python
data[~data['area'].isna()]['area'].head()
```

```python
data['area'] = data['area'].str.split('/').str[0].str.split('₹').str[1].str.replace(',', '').astype(float)
```

```python
data['area'].value_counts()
```

```python
data.info()
```

```python
data.rename(columns={'area':'price_per_sqft'}, inplace=True)
data.sample(3)
```

# bedRoom	bathroom	balcony cols

```python
data[data['bedRoom'].isnull()].sample(3)
```

```python
data = data[~data['bedRoom'].isnull()]
data.sample(3)
```

```python
data[data['bedRoom'].isnull()]
```

```python
data['bedRoom'].value_counts()
```

```python
data['bedRoom'].str.split(' ').str[0].value_counts()
```

```python
data[data['bedRoom']=='bedRoom']
```

```python
data = data[data['bedRoom']!='bedRoom']
data['bedRoom'].str.split(' ').str[0].value_counts()
```

```python
data['bedRoom'] = data['bedRoom'].str.split(' ').str[0].astype(int)
```

```python
data.info()
```

# bathroom

```python
data['bathroom'].str.split(' ').str[0].value_counts()
```

```python
data['bathroom'] = data['bathroom'].str.split(' ').str[0].astype(int)
```

# balcony

```python
data['balcony'].value_counts()
```

```python
data['balcony'].str.split(' ').str[0].str.strip().str.lower().replace('no', '0').value_counts()
```

```python
data['balcony'] = data['balcony'].str.split(' ').str[0].str.strip().str.lower().replace('no', '0')
```

```python
data.sample(3)
```

# additionalRoom col

```python
data['additionalRoom'].value_counts()
```

```python
data['additionalRoom'] = data['additionalRoom'].fillna('not available').str.strip().str.lower()
```

# floorNum col

```python
data['floorNum'].value_counts()
```

```python
data['floorNum'] = data['floorNum'].str.split(' ').str[0].str.lower()
```

```python
data['floorNum'].value_counts(dropna=False)
```

```python
data[data['floorNum'].isna()]
```

```python
data['floorNum'] = data['floorNum'].str.replace('ground', '0').str.replace('lower', '0').str.replace('basement', '-1').str.extract(r'(\d+)')
```

```python
data.info()
```

```python
data.sample(3)
```

# facing col

```python
data['facing'].str.lower().str.strip().value_counts(dropna=False)
```

```python
data['facing'] = data['facing'].str.lower().str.strip().fillna('NA')
```

# construcuting property_price col from price_per_sqft and rate cols

```python
data.sample(3)
```

```python
data.info()
```

```python
data['price_in_crores'] = data['price_in_crores'].astype(float)
```

```python
data.insert(loc=2, column='area_in_sqft', value=round((data['price_in_crores']*10000000)/data['price_per_sqft']))
```
