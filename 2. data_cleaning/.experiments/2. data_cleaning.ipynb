# price col

```python
data['price']
```

```python
data['price'].value_counts()
```

```python
data['price'] = data['price'].str.lower().str.strip()
data['price'].value_counts()
```

```python
data = data[data['price'] != 'price on request']
```

```python
data['price'].value_counts()
```

```python
def clean_price_col(row):
    if 'crore' in row:
        return round(float(row.split(' ')[0])/100, 2)
    if 'lac' in row:
        return round(float(row.split(' ')[0]), 2)
    else:
        return row
```

```python
data['price'].astype(str).str.lower().apply(clean_price_col).value_counts()
```

```python
data['price'] = data['price'].astype(str).str.lower().apply(clean_price_col)
```

```python
data.rename(columns={'price':'price in crores'}, inplace=True)
```

```python
data.sample(3)
```
